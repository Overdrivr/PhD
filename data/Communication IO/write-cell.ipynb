```python
import matplotlib.pyplot as plt
import numpy as np
import numpy as np
import matplotlib.pyplot as plt
import os
import collections
import pandas as pd
import matplotlib as mp
from matplotlib import gridspec
from matplotlib.patches import Rectangle

# Read simulation files
def datafileread(measurename,skipfirstrows):
    # Reading Datafiles
    path = measurename
    data = np.genfromtxt(path,
                        skip_header=skipfirstrows,
                        delimiter=' ',
                        dtype=(float,float),
                        unpack=True)
    return data

# measurement
time, enable_in, clock, rstb, word, en_out_1, en_out_8, w0, w1, w2, w3, w4, w5, w6, w7 = datafileread('write_cell_chronogram.csv',1)
time *= 1e6
```

```python
f, (ax1, ax2, ax3, ax4, ax5, ax6, ax7, ax8, ax9) = plt.subplots(9,1,figsize=(10,4.5), sharex=True)

ax1.plot(time, enable_in, 'g', label="en1")
ax2.plot(time, clock, 'g', label="clk")
ax3.plot(time, word, 'g',  label="bus")
ax4.plot(time, en_out_1,  label="en2")
ax5.plot(time, en_out_8,  label="en8")
ax6.plot(time, w0, 'r',   label="bit 0")
ax7.plot(time, w1, 'r',   label="bit 1")
ax8.plot(time, w2, 'r',   label="bit 2")
ax9.plot(time, w3, 'r',   label="bit 3")

ax1.get_yaxis().set_ticklabels([])
ax2.get_yaxis().set_ticklabels([])
ax3.get_yaxis().set_ticklabels([])
ax4.get_yaxis().set_ticklabels([])
ax5.get_yaxis().set_ticklabels([])
ax6.get_yaxis().set_ticklabels([])
ax6.get_yaxis().set_ticklabels([])
ax7.get_yaxis().set_ticklabels([])
ax8.get_yaxis().set_ticklabels([])
ax9.get_yaxis().set_ticklabels([])


yoff = 0.6
xoff = 1.4e1
xperiod = 1e1
w = [0, 1, 0, 1, 1, 0, 1, 1]

for i in range(len(w)):
    ax3.text(xoff + xperiod * i, yoff, str(w[i]),  fontsize=15)
    
#xoff = 1.5e-5
#for i in range(len(w)):
#    ax1.annotate("", xy=(xoff + xperiod*(i+1), 0), xytext=(xoff + xperiod*(i+1), 1),arrowprops=dict(arrowstyle="->"))


xoff = 2.1e1

ax6.text(xoff, yoff, str(w[0]),  fontsize=15)
ax7.text(xoff + xperiod, yoff, str(w[1]),  fontsize=15)
ax8.text(xoff + xperiod * 2, yoff, str(w[2]),  fontsize=15)
ax9.text(xoff + xperiod * 3, yoff, str(w[3]),  fontsize=15)
    
ax1.set_xlim([0, 1.2e2])

ax6.set_ylim([-0.5, 3])
ax8.set_ylim([-0.5, 3])

ax1.legend(loc="right")
ax2.legend(loc="right")
ax3.legend(loc="right")
ax4.legend(loc="right")
ax5.legend(loc="right")
ax6.legend(loc="right")
ax7.legend(loc="right")
ax8.legend(loc="right")
ax9.legend(loc="right")

plt.xlabel("Time (μs)")
plt.tight_layout()
f.subplots_adjust(hspace=0)
plt.savefig("../../src/3/figures/curve_write_cell.png", pad_inches=0.3)
plt.show()
```
