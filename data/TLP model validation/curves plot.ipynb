```python
import matplotlib.pyplot as plt
import numpy as np
import numpy as np
import matplotlib.pyplot as plt
import os
import collections
import matplotlib as mp
mp.style.use('classic')

# Read simulation files
def datafileread(measurename,skipfirstrows,delim='\t'):
    # Reading Datafiles
    path = measurename
    data = np.genfromtxt(path,
                        skip_header=skipfirstrows,
                        delimiter=delim,
                        dtype=(float,float),
                        unpack=True)
    return data
```

```python
# 10 V
## 25 ohms
time_mes_r25_10V, v_mes_r25_10V, i_mes_r25_10V = datafileread('mes_R25_10V.csv',1)
time_sim_r25_10V, v_sim_r25_10V, i_sim_r25_10V = datafileread('sim_R25_10V.csv',1,delim=',')
## open
time_mes_open_10V, v_mes_open_10V, i_mes_open_10V = datafileread('mes_open_10V.csv',1)
time_sim_open_10V, v_sim_open_10V, i_sim_open_10V = datafileread('sim_open_10V.csv',1,delim=',')
## short
time_mes_short_10V, v_mes_short_10V, i_mes_short_10V = datafileread('mes_short_10V.csv',1)
time_sim_short_10V, v_sim_short_10V, i_sim_short_10V = datafileread('sim_short_10V.csv',1,delim=',')

time_mes_short_10V *= 1e9
time_mes_open_10V *= 1e9
time_mes_r25_10V *= 1e9
```

```python
# 500 V
## 25 ohms
time_mes_r25_500V, v_mes_r25_500V, i_mes_r25_500V = datafileread('mes_R25_500V.csv',1)
time_sim_r25_500V, v_sim_r25_500V, i_sim_r25_500V = datafileread('sim_R25_500V.csv',1,delim=',')
## short
time_mes_short_500V, v_mes_short_500V, i_mes_short_500V = datafileread('mes_short_500V.csv',1)
time_sim_short_500V, v_sim_short_500V, i_sim_short_500V = datafileread('sim_short_500V.csv',1,delim=',')
## open
time_mes_open_500V, v_mes_open_500V, i_mes_open_500V = datafileread('mes_open_500V.csv',1)
time_sim_open_500V, v_sim_open_500V, i_sim_open_500V = datafileread('sim_open_500V.csv',1,delim=',')
## 1 nF
time_mes_1nF_500V, v_mes_1nF_500V, i_mes_1nF_500V = datafileread('mes_1nF_500V.csv',1)
time_sim_1nF_500V, v_sim_1nF_500V, i_sim_1nF_500V = datafileread('sim_1nF_500V.csv',1,delim=',')
## 470 nH
time_mes_470nH_500V, v_mes_470nH_500V, i_mes_470nH_500V = datafileread('mes_470nH_500V.csv',1)
time_sim_470nH_500V, v_sim_470nH_500V, i_sim_470nH_500V = datafileread('sim_470nH_500V.csv',1,delim=',')

time_mes_r25_500V *= 1e9
time_mes_short_500V *= 1e9
time_mes_open_500V *= 1e9
time_mes_1nF_500V *= 1e9
time_mes_470nH_500V *= 1e9
```

```python
# 1000 V
## 25 ohms
time_mes_r25_1000V, v_mes_r25_1000V, i_mes_r25_1000V = datafileread('mes_R25_1000V.csv',1)
time_sim_r25_1000V, v_sim_r25_1000V, i_sim_r25_1000V = datafileread('sim_R25_1000V.csv',1,delim=',')
## open
time_mes_open_1000V, v_mes_open_1000V, i_mes_open_1000V = datafileread('mes_open_1000V.csv',1)
time_sim_open_1000V, v_sim_open_1000V, i_sim_open_1000V = datafileread('sim_open_1000V.csv',1,delim=',')
## short
time_mes_short_1000V, v_mes_short_1000V, i_mes_short_1000V = datafileread('mes_short_1000V.csv',1)
time_sim_short_1000V, v_sim_short_1000V, i_sim_short_1000V = datafileread('sim_short_1000V.csv',1,delim=',')

time_mes_r25_1000V *= 1e9
time_mes_short_1000V *= 1e9
time_mes_open_1000V *= 1e9
```

## Simulation vs measurement - 25 ohms load at 10V

```python
#
f, (ax1, ax2) = plt.subplots(1,2,figsize=(10,4))
xlims = [-20, 280]
ax1.plot(time_mes_r25_10V, v_mes_r25_10V, label="measurement")
ax1.plot(time_sim_r25_10V, v_sim_r25_10V, label="simulation")
ax1.set_xlim(xlims)
#ax1.set_ylim([1, 13])
#ax1.set_title('TLP output characterization')
#ax1.get_xaxis().set_ticklabels([])
ax1.set_ylabel('Voltage (V)')
ax1.set_xlabel('Time (ns)')

ax2.plot(time_mes_r25_10V, i_mes_r25_10V, label="measurement")
ax2.plot(time_sim_r25_10V, i_sim_r25_10V, label="simulation")
ax2.set_xlim(xlims)
#ax2.set_ylim([-0.01, 0.01])
ax2.set_ylabel('current (A)')
ax2.set_xlabel('Time (ns)')

#
plt.tight_layout()
plt.legend(loc="best")
plt.tight_layout()
f.subplots_adjust(hspace=0)
plt.savefig("../../src/2/figures/tlp_comparison_R25_10V.png", pad_inches=0.3)
plt.show()
```
