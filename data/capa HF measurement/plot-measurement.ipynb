```python
import matplotlib.pyplot as plt
import numpy as np
import numpy as np
import matplotlib.pyplot as plt
import os
import collections
import matplotlib as mp
mp.style.use('classic')

# Read simulation files
def datafileread(measurename,skipfirstrows):
    # Reading Datafiles
    path = measurename
    data = np.genfromtxt(path,
                        skip_header=skipfirstrows,
                        delimiter='\t',
                        dtype=(float,float),
                        unpack=True)
    return data

# measurement
name = 'capa_mag_phase.csv'
freq, mag, phase = datafileread(name,1)

f, (ax1, ax2) = plt.subplots(2,1,figsize=(10,6))
#
ax1.plot(freq, mag, label="magnitude")
#ax1.set_title('magnitude')
ax1.set_xlabel('frequency (Hz)')
ax1.set_xscale('log')
ax1.set_yscale('log')
ax1.set_xlim([4e1, 1e8])
ax1.set_ylabel('magnitude (dB)')
ax1.legend(loc='best')
#
ax2.plot(freq, phase, 'r', label="phase")
#ax2.set_title('phase')
ax2.set_xscale('log')
ax2.set_xlim([4e1, 1e8])
ax2.set_xlabel('frequency (Hz)')
ax2.set_ylabel('Phase (deg)')
ax2.legend(loc='best')
#
plt.tight_layout()

plt.savefig("../../src/2/figures/capa_hf_response.png", pad_inches=0.3)
plt.show()
```
