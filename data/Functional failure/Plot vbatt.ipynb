```python
import matplotlib.pyplot as plt
import numpy as np
import numpy as np
import matplotlib.pyplot as plt
import os
import collections
import pandas as pd
import matplotlib as mp
from matplotlib import gridspec
mp.style.use('classic')

# Read simulation files
def datafileread(measurename,skipfirstrows):
    # Reading Datafiles
    path = measurename
    data = np.genfromtxt(path,
                        skip_header=skipfirstrows,
                        delimiter=',',
                        dtype=(float,float),
                        unpack=True)
    return data

# measurement
dummy, time, vboost, vclamp9, v1p0, v2p5a = datafileread('total_fullblock_m100v.csv',19)
dummy, dummy, dummy, time_mes, vboost_mes = datafileread('curve4_m90V_Ch1.csv',6)

time *= 1e6
time_mes = time_mes * 1e6 + 54

undersample = 100
time_mes = time_mes[::undersample]
vboost_mes = vboost_mes[::undersample]
```

```python
plt.figure(figsize=(10, 3.5)) 
plt.plot(time, vboost, label="simulation")
plt.plot(time_mes, vboost_mes, 'r-', label="measurement")
plt.xlabel("Time (Î¼s)")
plt.ylabel("Voltage (V)")
plt.xlim([53.5, 56])
plt.ylim([-4, 18])

plt.tight_layout()
plt.legend(loc="lower right")
plt.savefig("../../src/3/figures/vbatt.png", pad_inches=0.3)
plt.show()
```
