```python
import matplotlib.pyplot as plt
import numpy as np
import numpy as np
import matplotlib.pyplot as plt
import os
import collections
import matplotlib as mp

# Read simulation files
def numpyread(measurename):
    # Reading Datafiles
    path = measurename
    data = np.loadtxt(path,delimiter='\t')
    return data

# Read simulation files
def datafileread(measurename,skipfirstrows):
    # Reading Datafiles
    path = measurename
    data = np.genfromtxt(path,
                        skip_header=skipfirstrows,
                        delimiter='\t',
                        dtype=(float,float),
                        unpack=True)
    return data

# measurement
name = 'processing_rf_output.csv'
proc_time,proc_amp = numpyread(name)
name = 'reference_waveform.csv'
ref_time,ref_v, ref_i = datafileread(name,1)
name = 'output_integral.csv'
int_time,int_i = numpyread(name)

#
plt.figure(figsize=(12,6))
plt.xlim([-0.5e-7, 2.0e-7])
plt.plot(proc_time, proc_amp, label="frequency method")
plt.plot(ref_time, ref_i, label = "reference")
plt.plot(int_time, int_i, label="integration method")
plt.legend(loc="best")
plt.title('S11')
plt.xlabel('frequency (Hz)')
plt.ylabel('magnitude (dB)')

plt.tight_layout()
plt.savefig("../../src/3/figures/final_comparison_reconstructions.png", pad_inches=0.3)
plt.show()
```
