```python
import matplotlib.pyplot as plt
import numpy as np
import numpy as np
import matplotlib.pyplot as plt
import os
import collections
import matplotlib as mp
mp.style.use('classic')

# Read simulation files
def datafileread(measurename,skipfirstrows):
    # Reading Datafiles
    path = measurename
    data = np.genfromtxt(path,
                        skip_header=skipfirstrows,
                        delimiter=' ',
                        dtype=(float,float),
                        unpack=True)
    return data

#
time_total_m10, v_total_m10, i_total_m10 = datafileread('comparison_total_-10.csv',1)
time_model_m10, v_model_m10, i_model_m10 = datafileread('comparison_model_-10.csv',1)
#
time_total_20, v_total_20, i_total_20 = datafileread('comparison_total_20.csv',1)
time_model_20, v_model_20, i_model_20 = datafileread('comparison_model_20.csv',1)
#
time_total_40, v_total_40, i_total_40 = datafileread('comparison_total_40.csv',1)
time_model_40, v_model_40, i_model_40 = datafileread('comparison_model_40.csv',1)

time_total_m10 *= 1e6
time_model_m10 *= 1e6

time_total_20 *= 1e6
time_model_20 *= 1e6

time_total_40 *= 1e6
time_model_40 *= 1e6
```

## Model vs complete schematic at -10V

```python
#
f, (ax1, ax2) = plt.subplots(2,1,figsize=(10,4), sharex=True)
xlims = [9, 11]
ax1.plot(time_total_m10, v_total_m10, label="complete schematic")
ax1.plot(time_model_m10, v_model_m10, label="model")
ax1.set_xlim(xlims)
ax1.set_ylim([1, 13])
ax1.set_title('$V_{batt}$')
#ax1.get_xaxis().set_ticklabels([])
ax1.set_ylabel('voltage (V)')

ax2.plot(time_total_m10, i_total_m10 * 1e3, label="complete schematic")
ax2.plot(time_model_m10, i_model_m10 * 1e3, label="model")
ax2.set_xlim(xlims)
ax2.set_ylim([-20, 20])
ax2.set_ylabel('current (mA)')
ax2.set_xlabel('time (Î¼s)')

#
plt.tight_layout()
ax1.legend(loc="best")
#plt.tight_layout()
#f.subplots_adjust(hspace=0)
plt.savefig("../../src/4/figures/comparison_model_total_m10V.png", pad_inches=0.3)
plt.show()
```
